```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)
df.head()
```

## Player Count

* Display the total number of players


```python
total_player = df.groupby(['SN'])
len(total_player)

d={'Total Players':[len(total_player)]}
d1 = pd.DataFrame(data=d)
d1
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(df.groupby(['Item ID']))
avg_price = df['Price'].mean()
number_purchase = len(df)
revenue =df['Price'].sum()
d={'Number of Unique Items':[unique_items],'Average Price':[avg_price],'Number of Purchases':[number_purchase],
   'Total Revenue':[revenue]}
d1 = pd.DataFrame(data=d)
d1['Average Price'] = d1['Average Price'].map('${:,.2f}'.format)
d1['Total Revenue'] = d1['Total Revenue'].map('${:,.2f}'.format)
d1
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender1 =df[['SN','Gender']]
gender2 =gender1.drop_duplicates(subset=['SN'])
gender3 = pd.DataFrame(data=gender2.groupby(['Gender']).count())
gender4=gender3.rename(columns={'SN':'Total Count'})
gender4['Percentage of Players']=gender4['Total Count']/len(total_player)*100
gender4['Percentage of Players']=gender4['Percentage of Players'].map('{:,.2f}%'.format)
gender4
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gen_count = pd.DataFrame(data=df.groupby(['Gender']).count())
gen_count = gen_count[['Purchase ID']]
gen_count=gen_count.rename(columns={'Purchase ID':'Purchase Count'})
gen_mean = pd.DataFrame(data=df.groupby(['Gender']).mean())
gen_mean = gen_mean[['Price']]
gen_mean=gen_mean.rename(columns={'Price':'Average Purchase Price'})
gen_sum = pd.DataFrame(data=df.groupby(['Gender']).sum())
gen_sum = gen_sum[['Price']]
gen_sum=gen_sum.rename(columns={'Price':'Total Purchase Value'})
gen1 = pd.concat([gen_count,gen_mean],axis=1,sort=False)
gen2 = pd.concat([gen1,gen_sum],axis=1,sort=False)
gen3 = pd.concat([gen2,gender4],axis=1,sort=False)
gen3['Avg Total Purchase per Person']=gen3['Total Purchase Value']/gen3['Total Count']
gen4 = gen3[['Purchase Count','Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']]
gen4['Average Purchase Price']=gen4['Average Purchase Price'].map('${:,.2f}'.format)
gen4['Total Purchase Value']=gen4['Total Purchase Value'].map('${:,.2f}'.format)
gen4['Avg Total Purchase per Person']=gen4['Avg Total Purchase per Person'].map('${:,.2f}'.format)
gen4
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9.9,14.1,19.1,24.1,29.1,34.1,39.1,150]
group_names =['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

aged =df.drop_duplicates(subset=['SN'])
# aged = aged[['Purchase ID']]
aged['Age Group'] = pd.cut(aged['Age'], bins, labels=group_names)
aged = aged.groupby(['Age Group']).count()
# aged.groupby(['Age Group']).sum()
aged=aged[['Purchase ID']]
aged=aged.rename(columns={'Purchase ID':'Total Count'})
aged['Percentage of Players']=aged['Total Count']/len(total_player)*100
aged['Percentage of Players']=aged['Percentage of Players'].map('{:,.2f}%'.format)
aged
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0,9.9,14.1,19.1,24.1,29.1,34.1,39.1,150]
group_names =['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
df['Age Ranges'] = pd.cut(df['Age'], bins, labels=group_names)
age_count = pd.DataFrame(data=df.groupby(['Age Ranges']).count())
age_count = age_count[['Purchase ID']]
age_count=age_count.rename(columns={'Purchase ID':'Purchase Count'})
age_mean = pd.DataFrame(data=df.groupby(['Age Ranges']).mean())
age_mean = age_mean[['Price']]
age_mean=age_mean.rename(columns={'Price':'Average Purchase Price'})
age_sum = pd.DataFrame(data=df.groupby(['Age Ranges']).sum())
age_sum = age_sum[['Price']]
age_sum=age_sum.rename(columns={'Price':'Total Purchase Value'})

age1 = pd.concat([age_count,age_mean],axis=1,sort=False)
age2 = pd.concat([age1,age_sum],axis=1,sort=False)
age3 = age2
age4 = age3[['Purchase Count','Average Purchase Price','Total Purchase Value']]
age4['Avg Total Purchase per Person']=age4['Total Purchase Value']/aged['Total Count']
age4['Average Purchase Price']=age4['Average Purchase Price'].map('${:,.2f}'.format)
age4['Total Purchase Value']=age4['Total Purchase Value'].map('${:,.2f}'.format)
age4['Avg Total Purchase per Person']=age4['Avg Total Purchase per Person'].map('${:,.2f}'.format)
age4
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_count = pd.DataFrame(data=df.groupby(['SN']).count())
sn_count = sn_count[['Purchase ID']]
sn_count=sn_count.rename(columns={'Purchase ID':'Purchase Count'})
sn_mean = pd.DataFrame(data=df.groupby(['SN']).mean())
sn_mean = sn_mean[['Price']]
sn_mean=sn_mean.rename(columns={'Price':'Average Purchase Price'})
sn_sum = pd.DataFrame(data=df.groupby(['SN']).sum())
sn_sum = sn_sum[['Price']]
sn_sum=sn_sum.rename(columns={'Price':'Total Purchase Value'})
sn1 = pd.concat([sn_count,sn_mean],axis=1,sort=False)
sn2 = pd.concat([sn1,sn_sum],axis=1,sort=False)
sn4 =sn2.sort_values(by=['Total Purchase Value'], ascending=False)

sn4['Average Purchase Price']=sn4['Average Purchase Price'].map('${:,.2f}'.format)
sn4['Total Purchase Value']=sn4['Total Purchase Value'].map('${:,.2f}'.format)
sn4.sort_values(by=['Total Purchase Value'], ascending=False)
sn4
```
